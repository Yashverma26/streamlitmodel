# Label Encoding

```python
print(df['sex'].value_counts())
print(df['smoker'].value_counts())
print(df['region'].value_counts())
```

```python
df.head()
```

```python
# age - std scaler
# sex - label encoding
# bmi - std scaler
# children - normal
# smoker - label
# region - hot encoding but used label encoding 
# charges - target 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["sex"] = le.fit_transform(df["sex"])
```

```python
df.head()
```

```python
print(df['sex'].value_counts())
```

```python
df["smoker"] = le.fit_transform(df["smoker"])
```

```python
df.head()
```

```python
print(df['smoker'].value_counts())
```

```python
print(df['region'].value_counts())
```

```python
d = le.fit_transform(df["region"])
```

```python
dd = pd.DataFrame(d)
```

```python
dd.value_counts()
```

```python
df['region'] = d
```

```python
df.head()
```

# Modelling

```python
x = df[['age', 'sex', 'bmi','children', 'smoker','region']]
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
x_train
```

```python
y_train
```

```python
lm.predict(x_test)
```

```python
predictions = lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
print(predictions[0:5]) 
print(y_test[0:5])
```

```python
df_final = pd.DataFrame({'Predicted':predictions,'Actual':y_test})
```

```python
df_final = df_final.reset_index()
```

```python
df_final = df_final.drop(['index'],axis=1)
```

```python
fig= plt.figure(figsize=(16,8))
plt.plot(df_final[:50])
plt.legend(['Real','Prediction'])
```

```python
fig= plt.figure(figsize=(16,8))
sns.jointplot(x='Actual',y='Predicted',data=df_final,kind='reg',)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
